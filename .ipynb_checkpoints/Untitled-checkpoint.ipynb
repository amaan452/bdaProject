```r
### Task.2. Is there any Null Value present in any column ? Show with Heat-map.
```

```r
# Display the first 6 rows of the dataset
head(data)
```

```r
# Show where NA values are present
null_matrix <- is.na(data)

# Display the matrix of NA values
null_matrix
```

```r
# Count the number of NA values in each column
na_count <- colSums(is.na(data))

# Display the count of NA values for each column
na_count
```

```r
### seaborn library (heat-map)
```

```r
# Load required libraries
library(ggplot2)
library(reshape2)

# Create a matrix indicating missing values
missing_matrix <- is.na(data)

# Melt the matrix to long format for ggplot
missing_melted <- melt(missing_matrix)

# Create the heatmap
ggplot(missing_melted, aes(Var1, Var2)) +
  geom_tile(aes(fill = value), color = "white") +
  scale_fill_manual(values = c("FALSE" = "white", "TRUE" = "red")) +
  labs(x = "Columns", y = "Rows", fill = "Missing") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
### Q.2. In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.
```

```r
# Display the structure of the dataset to see data types
str(data)
```

```r
# Convert the 'Release_Date' column to Date format
data$Date_N <- as.Date(data$Release_Date, format = "%d/%m/%Y")


# If the dates are in a different format, adjust the 'format' string accordingly
```

```r
# Display the structure of the dataset to see data types
str(data)
```

```r
year_counts <- table(format(data$Date_N, "%Y"))
```

```r
year_counts
```

```r
# Extract the year from the 'Date_N' column
data$Year <- format(data$Date_N, "%Y")

# Create a bar plot of the year counts
ggplot(data, aes(x = Year)) +
  geom_bar(fill = "blue") +
  labs(title = "Year-wise Distribution", x = "Year", y = "Count") +
  theme_minimal()
```

```r
# Display the first 2 rows of the dataset
head(data, 2)
```

```r
library(dplyr)
```

```r
# Group by 'Category' and count occurrences
data %>%
  group_by(Category) %>%
  summarise(Count = n())
```

```r
# Create a count plot for the 'Category' column
ggplot(data, aes(x = Category)) +
  geom_bar(fill = "blue") +
  labs(title = "Count of Categories", x = "Category", y = "Count") +
  theme_minimal()
```

```r
#### Q.7. Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".
```

```r
head(data, 2)
```

```r
# Filter rows where Category is 'Movie' and Type is 'Comedies'
filtered_movies <- data[data$Category == 'Movie' & data$Type == 'Comedies', ]

# Display the filtered data nicely
library(knitr)  # Load the knitr package for nicer display
kable(filtered_movies)
```

```r
# Filter rows where Category is 'Movie' and Type is 'Comedies' or Country is 'United Kingdom'
filtered_data <- data[(data$Category == 'Movie' & data$Type == 'Comedies') | (data$Country == 'United Kingdom'), ]

# Display the filtered data nicely
library(knitr)  # Load the knitr package for nicer display
kable(filtered_data)
```

```r
### Q.8. In how many movies/shows, Tom Cruise was cast ?
```

```r
# Display the first 2 rows of the dataset
head(data, 2)
```

```r
# Filter rows where Cast is 'Tom Cruise'
filtered_cast <- data[data$Cast == 'Tom Cruise', ]

# Display the filtered data nicely
library(knitr)  # Load the knitr package for nicer display
kable(filtered_cast)
```

```r
# Drop rows with any missing values
data_new <- na.omit(data)
```

```r
# Filter rows where 'Cast' contains 'Tom Cruise'
filtered_cast <- data_new[grepl('Tom Cruise', data_new$Cast, ignore.case = TRUE), ]

# Display the filtered data nicely
library(knitr)  # Load the knitr package for nicer display
kable(filtered_cast)

```
